### 문자열 관련 파이썬 기초 문법

1. 문자열에 있는 특정 문자 갯수 세기 : count()
2. 문자열에 있는 특정 문자의 위치 파악 : index() -> 반드시 있어야 하는 값을 찾아와라!!
3. 문자열에 있는 특정 문자의 위치 파악 (에러를 발생 x) : find() -> -1
4. 서로 다른 문자열을 하나로 합치고자 할 때 : join()
5. 문자열에서 앞.뒤 공백 없앨 때 : strip() // rstrip() // lstrip() -> 인자값이 없으면 기본적으로 공백을 없앰
6. 대.소문자 변경 : upper(), lower()
7. 문자열을 쪼개고자 할 때 : split() -> 리스트를 반환 -> 인자값이 없으면 기본적으로 공백을 기준으로 나눔
8. 문자열을 찾아서 원하는 문자열로 교체하고자 할 때 : replace(인자값 => 대상, 무엇으로 교체할지)

```python
data = "Dave David"
data.count("D")
```

```python
string = "dave ID is Dave"
string.find("D")
```
