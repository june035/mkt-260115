# 파이썬 함수 (function = 기능)
- 함수 : 어떤 입력값을 받아서 내부에서 연산처리 후 해당 결과값을 외부로 출력할 수 있도록 기능하는 것
(프로그래밍 언어의 꽃)
- 함수에 값을 입력 & 출력 : input & output
- 함수가 생성(*선언) -> 함수가 존재 -> 값을 입력 (인자값 | 인수 -> 매개변수) -> 넣을 수도 있고, 없을 수도 있다!
- 함수를 사용(*호출)
- 함수의 선언부 & 호출부의 "인자값 | 인수 -> 매개변수"의 갯수는 기본적으로 같아야!!!
(어느 특정 한쪽이 상대적으로 갯수가 많거나 적으로 정상적으로 실행 x)
- return(문) : 함수의 내부연산 처리 후 특정 값을 함수 바깥으로 값을 보낼 수 있는 기능!

def func1() :
    실제 함수의 실행연산이 진행될 공간 -> 반복문, 조건문 사용!!!

# 파이썬 자료구조
- 문자열 = "string"
- 정수 = 10
- 실수 = 3.14
- 논리형 = True
> 단일값 저장 자료형태 1:1

- 리스트 = [1, 2, 3] -> 원하는 값을 추가.삭제 있음
- 튜플 = 리스트와 거의 똑같음 -> 최초에 선언된 형태의 자료구조를 변경할 수 없음! (1, 3, 5, 7) -> 데이터의 추가 혹은 삭제불가!
> iter & 반복문
> 리스트 & 튜플 안에 담겨있는 값들이 구체적으로 어떤 값을 의미하는지 알 수 없음!!

- 딕셔너리 = 사전 = {"apple": "사과"} -> "key / value" -> property = 속성
- 집합 = 수학시간에 배웠던 집합 개념
> 교집합, 합집합, 차집합, 공집합 등등
> 반복되어지는 값을 허용하지 않음!!
> 데이터를 크롤링 값을 찾아올 때, 반복적으로 입력된 값이 들어오는 경우 => 1회 처리
> 인스타그램 접속 > 데이터 크롤링 > 네일아트 > 사장님 > 똑같은 컨텐츠 반복

> 다중값 저장 자료형태 1:3 / 1:7

1. 리스트 : 변수명 = [값]
> 변수명 = list()

2. 튜플 : 변수명 = (값)
> 변수명 = tuple()

3. 딕셔너리 : 변수명 = {"" : ""}
> 변수명 = dict()

4. 세트 : 변수명 = {}
> 변수명 = set()

```python
def mul_return(a) :
    b = a + 1
    return a, b

result = mul_return(1)
print(result)
print(type(result)) # (a, b)
```

```python
def test_func1(a, b) :
    quot = a // b # 몫
    rem = a % b # 나머지
    return (quot, rem)
```

```python
# result = test_func1(10, 3)

# 구조분해할당 기능
(f, s) = test_func1(10, 3)

# (f, s) = (3, 1)
print(f, s)
```

```python
data1 = {"apple", "samsung", "lg"}
data2 = {"samsung", "xaiomi", "lg"}
```

```python
# 교집합
data1 & data2
```

```python
# 합집합
data1 | data2
```

```python
# 차집합
data1 - data2
```

```python
# 교집합 제외한 나머지 부분집합
data1 ^ data2
```

```python
# set 자료구조는 중복값을 허용하지 않는다!!!
data_list = ["apple", "dell", "samsung", "lg", "lg", "apple", "dell", "apple", "samsung"]
```

```python
data_list
```

```python
set_result = set(data_list) # 명시적 형변환 VS 자동 형변환
print(set_result)
```

```python
# 특정 사이트에서 데이터 크롤링 > 리스트 자료구조 > 중복 필연 > 세트 자료구조 > 리스트
```
