### 파이썬 requests를 활용한 크롤링 팁!
1) 파이썬 기초 문법
2) requests & bs4의 크롤링 기초 문법
3) 패턴학습
- 1)
- 단일 x // 복수의 값을 가지고 있는 부모 찾기 > 부모 찾고 자식 요소 찾기
- ul > li
- ul = unorder list = 정렬되지 않은 리스트 + li
- 2)
- 리스트 > 인덱스 & 슬라이싱


```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app")
soup = BeautifulSoup(res.text, "html.parser")

# print(soup)

# section = soup.select(".course.maintitle")
# print(section)

section = soup.find("ul", id="dev_course_list")
titles = section.find_all("li", "course")

# enumerate

for index, title in enumerate(titles) :
    print(index + 1, "댓글 수 :", title.get_text(" ", strip=True).split(" - ")[-1].replace("\n", " ").replace(" ", "")[-2:-1])
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app")
soup = BeautifulSoup(res.text, "html.parser")

items = soup.select("ul > li")
for item in items :
    print(item.get_text(" ", strip=True).replace("\n", "").replace(" ", ""))
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://crawlingtest.netlify.app")
soup = BeautifulSoup(res.text, "html.parser")

items = soup.select("#dev_course_list > .course")
for item in items :
    print(item.get_text(" ", strip=True))
```
